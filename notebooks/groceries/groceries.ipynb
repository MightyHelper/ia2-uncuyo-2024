```python
item_types = set(df["itemDescription"])
item_types
```

```python
item_type_occurrences = df["itemDescription"].value_counts()
item_type_occurrences
```

```python
# discard items with less than 100 occurrences
# item_type_occurrences = item_type_occurrences[item_type_occurrences > 100]

df = df[df["itemDescription"].isin(item_type_occurrences.index)]
```

```python
gb = df.groupby(["Member_number", "Date"])
df2 = df.copy()
df2["itemDescription"] = gb["itemDescription"].transform(lambda x: ";".join(set(x)))
df2 = df2.drop_duplicates(subset=["Member_number", "Date"])
df2.drop(columns=["Date", "Member_number"], inplace=True)
df2["itemDescription"] = df2["itemDescription"].apply(lambda x: x.split(";"))

df3 = pd.get_dummies(df["itemDescription"])
df3[["Member_number", "Date"]] = df[["Member_number", "Date"]]
df3 = df3.groupby(["Member_number", "Date"]).sum() > 0
```

```python
df3
```

```python
df3 = df3.reset_index().drop(columns=["Member_number", "Date"])
```

```python
from mlxtend.frequent_patterns import apriori
result = apriori(df3, min_support=0.01, use_colnames=True)
result
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(result, metric="confidence", min_threshold=0.1)
rules
```
